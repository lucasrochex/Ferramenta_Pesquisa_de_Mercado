```python
cidadesEstudadas = negocios['Cidade'].unique()
```

```python
# Inserido os estados
negocios['Estado'] = '0'
for cidade in cidadesEstudadas:
    print(cidade)
    estado = populacao.loc[populacao['Local']==cidade,'Estado'].iloc[0]
    negocios.loc[(negocios['Cidade']==cidade),'Estado'] = estado
```

```python
# Inserindo a população de fato
for cidade in cidadesEstudadas:
    estado = negocios.loc[negocios['Cidade']==cidade,'Estado'].iloc[0]
    populacaoCidade = populacao.loc[(populacao['Local']==cidade)&(populacao['Estado']==estado),'População'].values[0]
    negocios.loc[(negocios['Cidade']==cidade)&(negocios['Estado']==estado),'População'] = populacaoCidade
```

#### Parte 2: Inserindo o IDH de cada município

```python
import glob
```

```python
pasta = 'IDH'
arquivosIDH = glob.glob(pasta + '/*.csv')
```

```python
# Unindo todos arquivos de IDH
listaIDH = []
for arquivoIDH in arquivosIDH:
    dataframe = pd.read_csv(arquivoIDH)
    listaIDH.append(dataframe)
```

```python
idh = pd.concat(listaIDH,ignore_index = True)
```

#### Parte 3: Inserindo o PIB per capita

```python
pasta = 'PIB'
arquivosPIB = glob.glob(pasta + '/*.csv')
```

```python
# Unindo todos arquivos de PIB
listaPIB = []
for arquivoPIB in arquivosPIB:
    dataframe = pd.read_csv(arquivoPIB)
    listaPIB.append(dataframe)
```

```python
pib = pd.concat(listaPIB,ignore_index=True)
```

```python
negocios['pibPerCapita'] = ''
for i in range(len(negocios)):
    cidade = negocios['Cidade'][i]
    pibCidade = pib[(pib['Localidade'] ==cidade)&(pib['Posição']=='2.1')]['2018'].values[0]
    negocios.loc[i,'pibPerCapita'] = pibCidade
```

#### Parte 4: Inserindo frota de veículos

```python
pasta = 'Frota Veículos'
arquivosFrota = glob.glob(pasta + '/*.csv')
```

```python
listaFrota = []
for arquivoFrota in arquivosFrota:
    dataframe = pd.read_csv(arquivoFrota)
    listaFrota.append(dataframe)
```

```python
frota = pd.concat(listaFrota,ignore_index=True)
```

```python
negocios['Frota Total'] = ''
negocios['Frota Automóveis'] = ''
negocios['Frota Motos'] = ''
for i in range(len(negocios)):
    cidade = negocios['Cidade'][i]
    print(cidade)
    if len(frota[(frota['Localidade'] ==cidade)&(frota['Posição']=='1')]['2020'].values)>0:
        frotaTotal = frota[(frota['Localidade'] ==cidade)&(frota['Posição']=='1')]['2020'].values[0]
        negocios.loc[i,'Frota_Total'] = frotaTotal
        #Frota de automóveis
        frotaAuto = frota.loc[(frota['Localidade'] ==cidade)&(frota['Posição']=='1.1.1'),'2020'].values[0]
        negocios.loc[i, 'Frota_Automóveis'] = frotaAuto
        #Frota de Motos
        frotaMotos = frota.loc[(frota['Localidade'] ==cidade)&(frota['Posição']=='1.1.10'),'2020'].values[0]
        negocios.loc[i,'Frota_Motos'] = frotaMotos
    
```
